# analysis on bmi of people

```python
prd=df[df["Outcome"]==1]
```

```python
prd["grouped_BMI"]=pd.cut(prd.BMI, bins = [0,20,30,40,50])
```

```python
plt.figure(figsize=(10,5))

prd["grouped_BMI"].value_counts().plot(kind="pie")
plt.title("people with with diabetes and their bmi")
```

```python
plt.figure(figsize=(10,5))
prd["grouped_BMI"].value_counts().plot(kind="bar")
plt.title("people with with diabetes and their bmi")
```

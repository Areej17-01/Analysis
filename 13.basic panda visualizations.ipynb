```python
import pandas as pd 
```

```python
women=pd.read_csv("percent-bachelors-degrees-women-usa.csv")
```

```python
women.head()
```

```python
#.plot means blank graph
#value to be plotted on graph is tics
women.plot(kind="line",x="Year",y="Business")
```

```python
#rot to rotate
#in multiple plotting x-axis reamin same
women.plot(kind="line",x="Year",y=["Business","Computer Science"],grid=True,\
           xticks=range(1970,2020,5),yticks=range(0,101,10),color=["red","blue"],rot=90,figsize=(8,3)\
          )
```

```python
women.plot(kind="line",x="Year",y=["Business","Computer Science"],grid=True,\
           xticks=range(1970,2020,5),yticks=range(0,101,10),colormap="summer",rot=90,figsize=(8,3)\
          )
```

```python
#colormap=spectrals,gist_rainbow
y_axis=women.columns[1:]
women.plot(kind="line",x="Year",y=y_axis,grid=True,\
           xticks=range(1970,2020,5),yticks=range(0,101,10),colormap="summer",rot=90,figsize=(15,10)\
          ,fontsize=18)
```

```python
#bar chart
women.plot(kind="bar",x="Year",y="Business",yticks=range(0,101,5),figsize=(16,6))
```

```python
#bar chart
women.plot(kind="bar",x="Year",y=y_axis,yticks=range(0,101,5),figsize=(16,6))
```

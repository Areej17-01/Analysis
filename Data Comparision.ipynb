```python
#checking client names
```

```python
same_clients=0
list=[]
an_list=[]
for i in old_items["Client Name"]:
    for j in new_items["Client Name"]:
        if i in j:
            list.append(i)
            an_list.append(j)
            same_clients+=1
```

```python
for i,j in zip(an_list,list):
    temp=old_items[old_items["Client Name"]==j]
    temp["Client Name"]=i
    old_items[old_items["Client Name"]==j]=temp
```

```python
import pandas as pd
file=pd.read_csv("results.csv")
df={"Teams":None}
```

```python
df={"Teams":None}
df["Teams"]=file[file["season"]=="2006-2007"]["home_team"].unique()

list1=[]
def draw(coun): 
    
    data=file.loc[(file["season"]=="2006-2007")&((file["home_team"]==coun)|(file["away_team"]==coun))]
    list1.append(len(data[data["result"]=="D"]))
    
```

```python
for i in df["Teams"]:
    draw(i)
df["draw"]=list1
```

```python
limited_data=file.loc[(file["season"]=="2006-2007")]
```

```python
list1=[]
for i in df["Teams"]:
    list1.append(len(limited_data.loc[(limited_data["home_team"]==i) & (limited_data["result"]=="H")|(limited_data["away_team"]==i) & (limited_data["result"]=="A")]))
df["Wins"]=list1
```

```python
list1=[]
for i in df["Teams"]:
    list1.append(len(limited_data.loc[(limited_data["home_team"]==i) & (limited_data["result"]=="A")|(limited_data["away_team"]==i) & (limited_data["result"]=="H")]))
df["Loss"]=list1
```

```python
df=pd.DataFrame(df)
```

```python
limited_data.loc[(limited_data["home_team"]=="Arsenal")]["home_goals"].sum()+limited_data.loc[(limited_data["away_team"]=="Arsenal")]["away_goals"].sum()
limited_data.loc[(limited_data["home_team"]=="Arsenal")]["away_goals"].sum()+limited_data.loc[(limited_data["away_team"]=="Arsenal")]["home_goals"].sum()
```

```python
list1=[]

list2=[]
list3=[]
for i in df["Teams"]:
    list1.append(limited_data.loc[(limited_data["home_team"]==i)]["home_goals"].sum()+limited_data.loc[(limited_data["away_team"]==i)]["away_goals"].sum())
    list2.append(limited_data.loc[(limited_data["home_team"]==i)]["away_goals"].sum()+limited_data.loc[(limited_data["away_team"]==i)]["home_goals"].sum())
    list3.append((limited_data.loc[(limited_data["home_team"]==i)]["home_goals"].sum()+limited_data.loc[(limited_data["away_team"]==i)]["away_goals"].sum())-(limited_data.loc[(limited_data["home_team"]==i)]["away_goals"].sum()+limited_data.loc[(limited_data["away_team"]==i)]["home_goals"].sum()))
df["GF"]=list1
df["GA"]=list2
df["GD"]=list3
```

```python
df["points"]=((df["Wins"]*3)+(df["draw"]))
```

```python

df=df.sort_values(by=['points'], ascending=False)
```

```python

df=df.sort_values(by=['points'], ascending=False)

df.reset_index(inplace = True, drop = True)
df
```

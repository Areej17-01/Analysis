```python
def readfile(filename):
    from csv  import reader
    return list(reader(open(filename,encoding="utf-8")))
dataset=readfile("hacker_news.csv")
data=(dataset[1:])
import datetime as dt
```

```python
pattern="%m/%d/%Y %H:%M"
for i in data:
    i[-1]=dt.datetime.strptime(i[-1],pattern)
    print(i[-1])
```

```python
askhn=[]
showhn=[]
for i in data:
    if "Show HN:" in i[1]:
        askhn.append(i[1])
    if "Ask HN:" in i[1]:
        showhn.append(i[1])
```

```python
#full_record
ask_hn=[]
show_hn=[]
for i in data:
    if i[1].upper().startswith("ASK HN"):
        ask_hn.append(i)
    elif i[1].upper().startswith("SHOW HN"):
        show_hn.append(i)
```
